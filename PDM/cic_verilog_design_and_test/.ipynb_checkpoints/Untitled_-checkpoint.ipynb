## sine generation

```python
sample_rate = 4.8 * 10 ** 6
start_time = 0
end_time = 10

time = np.arange(start_time, end_time, 1/sample_rate)
frequency = 4.8 * 10 ** 3
amplitude = 1

sinewave = amplitude * np.sin(2 * np.pi * frequency * time)


amplitude2 = 2
sinewave2 = amplitude2 * np.sin(2 * np.pi * frequency * time)

amplitude3 = 10
sinewave3 = amplitude3 * np.sin(2 * np.pi * frequency * time)


sine = np.append(sinewave,sinewave2)
sine = np.append(sine, sinewave3)
```
